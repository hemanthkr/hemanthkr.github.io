## Activation Functions

An activation function is applied to the output of a neuron in a network. It introduces non-linearity which enables the network to model complex data distributions and learn features. There are many activation functions and the choice is always dependant on the type of problem and dataset available. We will only look at the most commonly used activation functions which should give us a good understanding.


(@) Rectified Linear Unit (ReLU) - Commonly used in hidden layers due to its simplicity and efficiency. Helps mitigate the vanishing gradient problem. Output range $[0, \infty)$

    $$f(x) = max(0, x)$$

(@) Hyperbolic Tangent (Tanh) - Often used in hidden layers when output needs to be centered around zero. Output range $(-1, 1)$

    $$f(x) = tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$

(@) Sigmoid - For binary classification output layers, sigmoid is commonly used. Squashes input values to the output range $(0, 1)$

    $$f(x) = \frac{1}{1 + e^{-x}}$$

(@) Softmax - Typically used in output layers for multi-class classification where the outputs are probabilities that sum to 1. Amplifies highest values while suppressing lower ones. Output range $(0, 1)$

    $$f(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n}e^{x_j}}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / exp_x.sum(axis=0)

# Generate input values
x = np.linspace(-10, 10, 100)

# Compute activation function outputs
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_softmax = softmax(x)  # Softmax is typically applied to vectors

# Plot the activation functions
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU")
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y_tanh, label="Tanh")
plt.title("Tanh")
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y_softmax, label="Softmax")
plt.title("Softmax")
plt.grid()

plt.tight_layout()
plt.show()
```

Activation functions are chosen based on the specific nature of the problem and the desired properties of the neural network.Some functions are more computationally efficient than others. In the next blog, we'll look at different types of loss functions.
